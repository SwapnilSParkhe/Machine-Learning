### 2) Requirements

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 3) Data Imports and Inspection

```python
df= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data', header=None)
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

### 4) Exploratory Data Analysis

**4.1) Univariate**

```python
df.describe(include='all').transpose()
```

```python
sns.countplot(x=1, data=df)
plt.show()
df[1].value_counts(normalize=True)*100
```

```python
df.hist(figsize=(16,20))
plt.show()
plt.tight_layout()
```

```python
df.boxplot(figsize=(16,4))

fig, axs = plt.subplots(4,8, figsize = (16, 10))
ax=axs.flatten()

for i, col in enumerate([col for col in df.columns if df[col].dtype!='object']):
    sns.boxplot(y=col, data=df, ax=ax[i])
    plt.tight_layout()
```

**4.2) Bivariate**

```python
fig, axs = plt.subplots(4,8, figsize = (16, 10))
ax=axs.flatten()

for i, col in enumerate([col for col in df.columns if df[col].dtype!='object']):
    sns.boxplot(x=1, y=col, data=df, ax=ax[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(df.corr(), cmap='plasma')
plt.title('Correlation among numeric vars')
plt.show()
```
